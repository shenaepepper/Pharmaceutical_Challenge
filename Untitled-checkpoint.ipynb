```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
import scipy.stats as st
from scipy.stats import linregress

# Study data files
Mouse_metadata_to_load = Path("Mouse_metadata.csv")
Study_results_to_load = Path("Study_results.csv")

# Read the mouse data and the study results
Mouse_metadata = pd.read_csv(Mouse_metadata_to_load)
Study_results = pd.read_csv(Study_results_to_load)
```

```python
# Merge the two DataFrames
merged_data = pd.merge(Mouse_metadata, Study_results, on='Mouse ID', how='outer')

# Display the merged DataFrame
print(merged_data)
```

```python
# Number of mice IDs
number_mice = merged_data['Mouse ID'].nunique()
print(f"Number of mice IDs: {number_mice}")

# Check for duplicate time points
duplicate_mouse_ids = merged_data[merged_data.duplicated(subset=['Mouse ID', 'Timepoint'])]
print("Mouse IDs with duplicate time points:")
print(duplicate_mouse_ids)

# Create a cleaned DataFrame by removing data for the mouse with duplicate time points
cleaned_data = merged_data[merged_data['Mouse ID'] != duplicate_mouse_ids.iloc[0]['Mouse ID']]
```

```python
# Updated number of mice IDs after removing the data for the mouse with duplicate time points
updated_number_mice = cleaned_data['Mouse ID'].nunique()
print(f"Updated number of mice IDs: {updated_number_mice}")
```

```python
# Group the merged dataframe by 'Drug Regimen'
grouped_data = merged_data.groupby('Drug Regimen')

# Calculate the summary statistics for tumor volume within each regimen group
summary_statistics = grouped_data['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

# Rename the columns for clarity
summary_statistics = summary_statistics.rename(columns={
    'mean': 'Mean Tumor Volume',
    'median': 'Median Tumor Volume',
    'var': 'Tumor Volume Variance',
    'std': 'Tumor Volume Std. Deviation',
    'sem': 'Tumor Volume SEM'
})

# Display the summary statistics DataFrame
print(summary_statistics)
```

```python
# Calculate the count of rows (Mouse ID/Timepoints) in each group
count_by_regimen = grouped_data['Mouse ID'].count()

# Create a bar chart using Pandas DataFrame.plot() method
count_by_regimen.plot(kind='bar')
plt.xlabel('Drug Regimen')
plt.ylabel('Total Number of Observed Mouse Timepoints')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group the data by 'Sex' and calculate the count of mice in each group
sex_distribution = merged_data['Sex'].value_counts()

# Create a pie chart using Pandas DataFrame.plot() method
sex_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=360)
plt.title('Distribution of Mice by Sex')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
# Set the y-axis label
plt.ylabel('Sex')
plt.show()
```

```python
# List of the four most promising treatments
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create a grouped DataFrame to find the last time point for each mouse
max_timepoint_df = merged_data.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this grouped DataFrame with the original cleaned DataFrame to get the final tumor volumes
final_tumor_volumes = pd.merge(max_timepoint_df, merged_data, on=['Mouse ID', 'Timepoint'], how='left')

# Filter the final tumor volumes DataFrame to only include the four specified treatments
final_tumor_volumes = final_tumor_volumes[final_tumor_volumes['Drug Regimen'].isin(treatments)]

# Create a list to hold treatment names
treatment_names = treatments

# Create an empty list to hold the tumor volume data
tumor_volume_data = []

# Loop through each treatment and append the corresponding tumor volume data
for treatment in treatments:
     # Filter the final tumor volumes DataFrame for the current treatment
    treatment_df = final_tumor_volumes[final_tumor_volumes['Drug Regimen'] == treatment]
    
    # Append the resulting final tumor volumes for the current treatment to the list
    tumor_volume_data.append(treatment_df['Tumor Volume (mm3)'])
    
# Determine potential outliers and print the results
outliers = []

for i, treatment in enumerate(treatments):
    quartiles = tumor_volume_data[i].quantile([0.25, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
    lower_bound = lower_q - 1.5 * iqr
    upper_bound = upper_q + 1.5 * iqr
    
    # Find potential outliers and append them to the outliers list
    potential_outliers = tumor_volume_data[i][(tumor_volume_data[i] < lower_bound) | (tumor_volume_data[i] > upper_bound)]
    outliers.extend(potential_outliers)
 
 # Print the results
    print(f"Potential Outliers for {treatment}:")
    print(potential_outliers)

```

```python
# List of the four most promising treatments
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create a grouped DataFrame to find the last time point for each mouse
max_timepoint_df = merged_data.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this grouped DataFrame with the original cleaned DataFrame to get the final tumor volumes
final_tumor_volumes = pd.merge(max_timepoint_df, merged_data, on=['Mouse ID', 'Timepoint'], how='left')

# Create an empty list to hold the final tumor volumes for each drug
tumor_volume_data = []

# Loop through each treatment
for treatment in treatments:
    # Filter the final tumor volumes DataFrame for the current treatment
    treatment_df = final_tumor_volumes[final_tumor_volumes['Drug Regimen'] == treatment]
    
    # Append the resulting final tumor volumes for the current treatment to the list
    tumor_volume_data.append(treatment_df['Tumor Volume (mm3)'])

# Create a figure and axis for the box plot
fig, ax = plt.subplots()

# Generate a box plot for the final tumor volumes of all treatments
boxprops = dict(linestyle='-', linewidth=2, color='blue', facecolor='lightblue')
medianprops = dict(linestyle='-', linewidth=2, color='red')
flierprops = dict(marker='o', markersize=12, markerfacecolor='red', markeredgecolor='black')

# Plot the box plot
bp = ax.boxplot(tumor_volume_data, patch_artist=True, boxprops=boxprops, medianprops=medianprops, flierprops=flierprops)

# Set labels and title
ax.set_xlabel('Treatment Regimen')
ax.set_ylabel('Final Tumor Volume (mm3)')  # Set y-axis label
ax.set_title('Final Tumor Volume Distribution by Treatment Group')

# Set x-axis labels
ax.set_xticklabels(treatments)

# Highlight potential outliers
for flier in bp['fliers']:
    flier.set(marker='.', color='red', markersize=8, markeredgewidth=1)
        
# Show the box plot
plt.show()
```

```python
# Filter the data for the Capomulin treatment regimen
capomulin_data = merged_data[merged_data['Drug Regimen'] == 'Capomulin']

# Calculate the average tumor volume for each mouse
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().reset_index()

# Group the data by 'Mouse ID' to obtain the weight of each mouse
mouse_weight = capomulin_data.groupby('Mouse ID')['Weight (g)'].max().reset_index()

# Merge the dataframes on 'Mouse ID' to combine weight and average tumor volume data
mouse_data = pd.merge(average_tumor_volume, mouse_weight, on='Mouse ID')
# Calculate the correlation coefficient
correlation = mouse_data['Weight (g)'].corr(mouse_data['Tumor Volume (mm3)'])

# Create a scatter plot of mouse weight versus average tumor volume
plt.figure(figsize=(8, 6))
plt.scatter(mouse_data['Weight (g)'], mouse_data['Tumor Volume (mm3)'], marker='o', facecolors='blue', edgecolors='black')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Treatment Regimen)')
plt.grid(True)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(mouse_data['Weight (g)'], mouse_data['Tumor Volume (mm3)'])

# Create the regression line equation
line_eq = f"y = {slope:.2f}x + {intercept:.2f}"

# Calculate the regression line values
regression_values = slope * mouse_data['Weight (g)'] + intercept

# Plot the regression line
plt.plot(mouse_data['Weight (g)'], regression_values, color='red', linestyle='--', label=line_eq)

# Display the correlation coefficient
plt.legend([f'Correlation: {correlation:.2f}'])
plt.show()

# Output the correlation coefficient and linear regression equation
print(f"Correlation Coefficient: {correlation:.2f}")
print(f"Linear Regression Equation: {line_eq}")
```
